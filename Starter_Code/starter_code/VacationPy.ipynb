### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
# Load continent boundaries from the Natural Earth dataset
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))


# Create a GeoDataFrame using  latitude and longitude
gdf_cities = gpd.GeoDataFrame(
    city_data_df, 
    geometry=gpd.points_from_xy(city_data_df['Lng'], city_data_df['Lat']),
    crs='EPSG:4326'
)

# Set up Seaborn style
sns.set_style("darkgrid")

# Create subplots
fig, ax = plt.subplots(figsize=(12, 8))

# Plot world boundaries
world.boundary.plot(ax=ax, linewidth=1, color='black')

# Determine the range of humidity values
min_humidity = city_data_df['Humidity'].min()
max_humidity = city_data_df['Humidity'].max()

# Create a scatter plot with dynamic hover labels and adjusted color scale
scatter = sns.scatterplot(
    data=gdf_cities,
    x='Lng',
    y='Lat',
    hue='Humidity',
    palette='coolwarm',
    size='Humidity',
    sizes=(10, 100),  
    alpha=0.7,
    ax=ax,
    vmin=min_humidity, 
    vmax=max_humidity ,
    legend="brief"
)

# Create a colorbar for humidity
cbar = plt.colorbar(scatter.collections[0], ax=ax)
cbar.set_label('Humidity')

# Set axis labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('World Map with City Humidity')


# Show the plot
plt.show()
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Narrow down cities that fit criteria and drop any results with null values

humid_condition = city_data_df[(city_data_df['Humidity'] >50 ) & (city_data_df['Humidity']<100)]  
temp_condition = humid_condition[(humid_condition['Max Temp'] > 290)]
filtered_city_df = temp_condition[temp_condition['Cloudiness']==0]

# Drop any rows with null values
city_df_cleaned = filtered_city_df.dropna(how="any")

# # Display sample data
city_df_cleaned.head()
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = city_df_cleaned[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df['Hotel Name'] = ''

# Display sample data
hotel_df.head()
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
# Set parameters to search for a hotel
radius = 10000
params = {
   'type': 'accommodation.hotel',
   'apiKey': geoapify_key
}

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # get latitude, longitude from the DataFrame
    lat = row['Lat']
    lng = row['Lng']
    
    # Add filter and bias parameters with the current city's latitude and longitude to the params dictionary
    params["filter"] = f"circle:{lng},{lat},{radius}"
    params["bias"] = f"proximity:{lng},{lat}"

    
    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"


    # Make and API request using the params dictionaty
    name_address = requests.get(base_url, params=params)
    
    # Convert the API response to JSON format
    name_address = name_address.json()
    
    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"
        
    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df.head()
```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
import geopandas as gpd
import geoviews as gv
import pandas as pd
import holoviews as hv
from bokeh.models import HoverTool

# Load continent boundaries from the Natural Earth dataset
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))


# Create a GeoDataFrame from your DataFrame using latitude and longitude
gdf_cities = gpd.GeoDataFrame(
    hotel_df,
    geometry=gpd.points_from_xy(hotel_df['Lng'], hotel_df['Lat']),
    crs='EPSG:4326'
)

# Define a GeoDataFrame to hold the world map data
world_map = gv.Polygons(world, vdims=['name'])

# Define a GeoDataFrame for the city data with point size and color based on humidity
city_points = gv.Points(gdf_cities, vdims=['City', 'Humidity', 'Hotel Name', 'Country']).opts(
    size=gv.dim('Humidity')*.1, 
    color='Humidity', 
    cmap='viridis', 
    tools=['hover'],
)

# Combine the world map with the city data and set legends
world_plot = (world_map * city_points).opts(
    width=1200,  
    height=800,  
    title='World Map with Hover functionality',
    show_legend=True,
)

# Customize the hover tooltip using Bokeh
hover = HoverTool()
hover.tooltips = [
    ("City", "@City"),
    ("Humidity", "@Humidity"),
    ("Hotel Name", "@{Hotel Name}"),
    ("Country", "@Country"),
]

# Adding the hover tool to the plot
world_plot = world_plot.opts(tools=[hover])

%output holomap='bokeh'
world_plot
```
